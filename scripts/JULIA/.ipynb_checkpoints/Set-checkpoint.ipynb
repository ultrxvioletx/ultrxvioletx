```julia
module CardClass
export Card, print
    struct Card
        number::Int64
        color::Char
        figure::Char
        texture::Char
    end
number(c::Card) = c.number
color(c::Card) = c.color
figure(c::Card) = c.figure
texture(c::Card) = c.texture
end #module
```

```julia
numbers = [1,2,3]
colors = ['R','P','G']
figures = ['*','+','#']
textures = [Char(0x25A0),Char(0x25A1),Char(0x25A5)] #filled, empty, lined
```

```julia
function isEqual(arr)
    eq = all(arr[1] .== arr)
    diff = (length(Set(arr)) == length(arr))
    return (eq || diff, eq)
end

function isSet(hand)
    nums = [card.number for card in hand]
    cols = [card.color for card in hand]
    texs = [card.texture for card in hand]
    figs = [card.figure for card in hand]

    flag = true
    counter = 0
    for arr in [nums,cols,texs,figs]
        flag = flag && isEqual(arr)[1]
        if flag
            if isEqual(arr)[2]
                counter += 1
            end
        else
            break
        end
    end
    return (flag, counter)
end
```

```julia
function getComplement(c1,c2)
    c3 = []
    fields = fieldnames(Card)
    for (i,arr) in pairs([numbers,colors,textures,figures])
        if getfield(c1,fields[i]) == getfield(c2,fields[i])
            append!(c3, getfield(c1,fields[i]))
        else
            append!(c3, setdiff(arr,[getfield(c1,fields[i]),getfield(c2,fields[i])])[1])
        end
    end
    return Card(c3[1],c3[2],c3[3],c3[4])
end
```

```julia
function findSet()
    for i in (1:1_000)
        n = sample(1:81, 3; replace=false)
        hand = [deck[n[1]],deck[n[2]],deck[n[3]]]
        flag,counter = isSet(hand)
        if flag
            return (i,counter,hand)
            break
        end
    end
end


using Plots
using StatsBase
using .CardClass

deck = [Card(i,j,k,l) for i in numbers, j in colors, k in figures, l in texture]
hist_data = zeros(Int, 4)
for i in 1:100_000
    intentos,equals,hand = findSet()
    hist_data[equals + 1] += 1
end
plt = bar(0:3, hist_data, xlabel="Equals", ylabel="Frecuency", legend=false)
display(plt)
```

```julia
function searchSet(board)
    b = length(board)
    set = []
    for i in 1:b-2, j in i+1:b-1, k in j+1:b
        c1 = board[i]
        c2 = board[j]
        c3 = board[k]
        (isset,counter) = isSet([c1,c2,c3])
        if isset
            set = [c1,c2,c3]
            @show set
            return (set,counter)
            break
        end
    end
end


using StatsBase
using .CardClass


deck = [Card(i,j,k,l) for i in numbers, j in colors, k in figures, l in texture]
n = sample(1:length(deck), 12; replace=false)
board = [deck[i] for i in n]
sets = []
hist_data = zeros(Int, 4)


while length(deck)>0
    deck = setdiff(deck,board)
    (set,counter) = searchSet(board)
    board = setdiff(board,set)
    b = length(board)
    d = length(deck)
    if b == 9 || b == 12
        n = sample(1:d, 3; replace=false)
        append!(board,[deck[i] for i in n])
        if !isempty(set)
            append!(sets,set)
            hist_data[counter + 1] += 1
        end
    elseif b == 15
        append!(deck,board)
        n = sample(1:d, 12; replace=false)
        board = [deck[i] for i in n]
        println("not found")
    end
end
```
